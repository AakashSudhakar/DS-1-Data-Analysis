```python
# kNN Algorithm for Extra Analysis
class k_Nearest_Neighbors_Algorithm(object):
    
    def __init__(self):
        # TODO: IMPORT MAIN DATA FILE HERE:
        self.FILE = open('./datasets/MSSA_dataset.csv')
    
    def convert_file_to_matrix(self):
        classifier_dictionary = {'promoter': 3, 'passive': 2, 'detractor': 1}
        
        arr_of_lines = self.FILE.readlines()
        num_of_lines = len(arr_of_lines)
        
        dataset = np.zeros((num_of_lines, 3))
        class_label_vector = []
        index = 0
        
        for line in arr_of_lines:
            line = line.strip()
            list_from_line = line.split('\t')
            dataset[index, :] = list_from_line[0:3]
            
            if (list_from_line[-1].isdigit()):
                class_label_vector.append(int(list_from_line[-1]))
            else:
                class_label_vector.append(classifier_dictionary.get(list_from_line[-1]))
            index += 1
            
        return dataset, class_label_vector
    
    def basic_label_classifier(self, inX, dataset, labels, k):
        dataset_size = dataset.shape[0]
        
        diff_mat = np.tile(inX, (dataset_size, 1)) - dataset
        sq_diff_mat = diff_mat ** 2
        sq_distances = sq_diff_mat.sum(axis = 1)
        distances = sq_distances ** 0.5
        
        sorted_dist_indices = distances.argsort()
        class_count = dict()
        
        for iterator in range(k):
            voting_label = labels[sorted_dist_indices[iterator]]
            class_count[voting_label] = class_count.get(voting_label, 0) + 1
            
        sorted_class_count = sorted(class_count.items(), key = op.itemgetter(1), reverse = True)
        return sorted_class_count[0][0]
    
    def auto_linear_normalization(self, dataset):
        min_vals = dataset.min(0)
        max_vals = dataset.max(0)
        ranges = max_vals - min_vals
        
        sample_dataset = dataset.shape[0]
        norm_dataset = np.zeros(np.shape(dataset))
        norm_dataset = dataset - np.tile(min_vals, (sample_dataset, 1))
        norm_dataset = norm_dataset / np.tile(ranges, (sample_dataset, 1))
        
        return norm_dataset, ranges, min_vals, max_vals
    
    def predictive_nps(self):
        nps_prediction_vector = ['DET', 'PAS', 'PRO']
        
        attr_location = str(input('\nWhere would you be interested in attending Summer Academy? >> '))
        attr_challenge = float(input('Do you enjoy more relaxed work or challenging work? >> '))
        attr_track = str(input('What track are you interested in? >> '))
        
        # IMPORT COMPLETE DATASET 2016-2017 HERE:
        nps_dataset, nps_labels = self.convert_file_to_matrix()
        norm_dataset, ranges, min_vals, max_vals = self.auto_linear_normalization(nps_dataset)
        
        attr_arr = np.array([attr_location, attr_track, attr_challenge])
        classifier_response = self.basic_label_classifier((attr_arr - min_vals) / ranges, norm_dataset, nps_labels, 3)
                
        # ENTER OUTPUT INFORMATION HERE!
        return print('\nThis individual is most likely a... {}!'.format(classifier_response - 1))
```

```python
# Create instance and run kNN against Make School SA data

kNN = k_Nearest_Neighbors_Algorithm()
kNN.predictive_nps()
```
