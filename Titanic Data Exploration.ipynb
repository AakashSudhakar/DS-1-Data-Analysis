```python
# Let's find out how many possible Jacks and Roses there are on the Titanic! (Inspired by Kaggle dataset)

possible_Jacks = titanic_df[(titanic_df["Sex"] == "male") & (titanic_df["Age"] == 20) & (titanic_df["Pclass"] == 3) & (titanic_df["Survived"] == 0)]
possible_Roses = titanic_df[(titanic_df["Sex"] == "female") & (titanic_df["Age"] == 17) & (titanic_df["Pclass"] == 1) & (titanic_df["Survived"] == 1)]

# possible_Jacks
# possible_Roses

labels = "Possible Jacks", "Possible Roses"
sizes = [10, 2]
explode = (0, 0.1)

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct="%i%%", shadow=True, startangle=90)
ax.axis("equal")
plt.title("How many possible Jacks and Roses were on board the Titanic?")
plt.show()
```

```python
# Q.1: Given that a passenger embarked at Queenstown, did they have a family? Who survived? Who died? 
# Eq.1: P(SibSp >= 1 OR ParCh >= 1 | Embarked == Q)

queenstown_df = titanic_df[titanic_df["Embarked"] == "Q"]
# queenstown_df.describe()

queenstown_survived_df = queenstown_df[queenstown_df["Survived"] == 1]
queenstown_died_df = queenstown_df[queenstown_df["Survived"] != 1]
# queenstown_survived_df.describe()

queenstown_family_df = queenstown_df[(queenstown_df["SibSp"] >= 1) | (queenstown_df["Parch"] >= 1)]
# queenstown_family_df.describe()

queenstown_family_survived_df = queenstown_family_df[queenstown_family_df["Survived"] == 1]
queenstown_family_died_df = queenstown_family_df[queenstown_family_df["Survived"] != 1]
# queenstown_family_died_df.describe()
```

```python
# Q.2: Given that a passenger embarked at Southampton, did they have a family? Who survived? Who died? 
# Eq.2: P(SibSp >= 1 OR ParCh >= 1 | Embarked == S)

southampton_df = titanic_df[titanic_df["Embarked"] == "S"]
# southampton_df.describe()

southampton_survived_df = southampton_df[southampton_df["Survived"] == 1]
southampton_died_df = southampton_df[southampton_df["Survived"] != 1]
# southampton_died_df.describe()

southampton_family_df = southampton_df[(southampton_df["SibSp"] >= 1) | (southampton_df["Parch"] >= 1)]
# southampton_family_df.describe()

southampton_family_survived_df = southampton_family_df[southampton_family_df["Survived"] == 1]
southampton_family_died_df = southampton_family_df[southampton_family_df["Survived"] != 1]
# southampton_family_died_df.describe()
```

```python
# Q.3: Given that a passenger embarked at Cherbourg, did they have a family? Who survived? Who died? 
# Eq.3: P(SibSp >= 1 OR ParCh >= 1 | Embarked == C)

cherbourg_df = titanic_df[titanic_df["Embarked"] == "C"]
# cherbourg_df.describe()

cherbourg_survived_df = cherbourg_df[cherbourg_df["Survived"] == 1]
cherbourg_died_df = cherbourg_df[cherbourg_df["Survived"] != 1]
# cherbourg_died_df.describe()

cherbourg_family_df = cherbourg_df[(cherbourg_df["SibSp"] >= 1) | (cherbourg_df["Parch"] >= 1)]
# cherbourg_family_df.describe()

cherbourg_family_survived_df = cherbourg_family_df[cherbourg_family_df["Survived"] == 1]
cherbourg_family_died_df = cherbourg_family_df[cherbourg_family_df["Survived"] != 1]
cherbourg_family_survived_df.describe()
```

```python
# Q.1-3: Plotting familial embarking data side-by-side in plot

index = np.arange(3)
WIDTH = 0.1

fig = plt.figure()
ax = fig.add_subplot(111)

data_embarked_family = np.array([20, 251, 83])
data_emb_fam_surv = np.array([7, 116, 56])
data_emb_fam_died = np.array([13, 135, 27])
data_embarked_all = np.array([77, 644, 168])
data_emb_all_surv = np.array([30, 217, 93])
data_emb_all_died = np.array([47, 427, 75])

rects_emb_fam = ax.bar(index, data_embarked_family, WIDTH, color="#3cab23")
rects_emb_fam_died = ax.bar(index + WIDTH, data_emb_fam_died, WIDTH, color="#1f5211")
rects_emb_fam_surv = ax.bar(index + WIDTH, data_emb_fam_surv, WIDTH, bottom = data_emb_fam_died, color="#5dff35")
rects_emb_all = ax.bar(index + 0.05 + WIDTH * 2, data_embarked_all, WIDTH, color="#5185ff")
rects_emb_all_died = ax.bar(index + 0.05 + WIDTH * 3, data_emb_all_died, WIDTH, color="#374da8")
rects_emb_all_surv = ax.bar(index + 0.05 + WIDTH * 3, data_emb_all_surv, WIDTH, bottom = data_emb_all_died, color="#91a7ff")

def make_bar_label(rects):
    """Function to automatically create labels for bar height"""
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2., 1.01 * h, "{}".format(h), ha = "center", va = "bottom")
        
make_bar_label(rects_emb_fam)
make_bar_label(rects_emb_all)

ax.set_xticks(index + WIDTH + 0.075)
ax.set_xticklabels(("Queenstown", "Southampton", "Cherbourg"))
ax.set_ylabel("Number of Passengers")
ax.set_ylim(0, 700)
ax.legend((rects_emb_fam[0], rects_emb_all[0]), ("Passengers w/ Families", "All Passengers"))
ax.set_title("\nComparing Passenger Survival Across Embarking Points with/without Families\n")

plt.show()
```

```python
# Q.4-9: Given that a passenger is from Queenstown/Southampton/Cherbourg, what is the chance that they survived/died? 

queenstown_df = titanic_df[titanic_df["Embarked"] == "Q"]
# queenstown_df.describe()

queenstown_survived_df = queenstown_df[queenstown_df["Survived"] == 1]
queenstown_died_df = queenstown_df[queenstown_df["Survived"] != 1]
# queenstown_survived_df.describe()

southampton_df = titanic_df[titanic_df["Embarked"] == "S"]
# southampton_df.describe()

southampton_survived_df = southampton_df[southampton_df["Survived"] == 1]
southampton_died_df = southampton_df[southampton_df["Survived"] != 1]
# southampton_survived_df.describe()

cherbourg_df = titanic_df[titanic_df["Embarked"] == "C"]
# cherbourg_df.describe()

cherbourg_survived_df = cherbourg_df[cherbourg_df["Survived"] == 1]
cherbourg_died_df = cherbourg_df[cherbourg_df["Survived"] != 1]
# cherbourg_survived_df.describe()


index = np.arange(3)
WIDTH = 0.1

fig = plt.figure()
ax = fig.add_subplot(111)

data_embarked_all = np.array([77, 644, 168])
data_emb_all_surv = np.array([30, 217, 93])
data_emb_all_died = np.array([47, 427, 75])

blue_scat = plt.scatter(index, data_embarked_all)
blue_plot = plt.plot(index, data_embarked_all)
plt.fill_between(index, data_embarked_all, color="skyblue")

red_scat = plt.scatter(index, data_emb_all_died)
red_plot = plt.plot(index, data_emb_all_died)
plt.fill_between(index, data_emb_all_died, color="red")

green_scat = plt.scatter(index, data_emb_all_surv)
green_plot = plt.plot(index, data_emb_all_surv)
plt.fill_between(index, data_emb_all_surv, color="lime")

blue_patch = mpatches.Patch(color="skyblue", label='All passengers')
red_patch = mpatches.Patch(color='red', label='Passengers that died')
green_patch = mpatches.Patch(color='lime', label='Passengers that lived')
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc = 2, ncol = 3, borderaxespad=0., handles=[blue_patch, red_patch, green_patch])
plt.title("\nComparing Passenger Survival Across Embarking Points with/without Families\n\n")

# Catch case for anomalous red-green Cherbourg crossover
# NOTE: remove red line that is under green curve

ax.set_ybound(0, 670)
ax.set_xticks(index + WIDTH - 0.1)
ax.set_xticklabels(("Queenstown", "Southampton", "Cherbourg"))
ax.set_ylabel("Number of Passengers")
plt.show()
```
